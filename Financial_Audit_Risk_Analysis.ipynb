## 1. Data Preparation
Simulating and structuring financial records for analysis.

```python
import plotly.express as px
import pandas as pd
import numpy as np

# --- Data Initialization & Preprocessing ---
# Simulating financial dataset for risk assessment validation
data = {
    'Department': ['IT', 'Marketing', 'Sales', 'Logistics', 'Legal', 'IT', 'Marketing', 'Sales', 'Logistics', 'Legal'],
    'Expense_Type': ['Software', 'Advertising', 'Travel', 'Fuel', 'Consulting', 'Hardware', 'Events', 'Bonus', 'Maintenance', 'Fines'],
    'Amount_KZT': [1500000, 2800000, 450000, 890000, 1200000, 3200000, 1100000, 200000, 600000, 150000],
    'Risk_Score': [0.1, 0.85, 0.2, 0.4, 0.95, 0.15, 0.6, 0.05, 0.3, 0.9] 
}

df = pd.DataFrame(data)

# --- Visualization Logic ---
# Implementing multi-dimensional scatter plot to analyze financial anomalies
# Size maps to transaction volume; Color maps to risk coefficient
fig = px.scatter(df, 
                 x="Department", 
                 y="Amount_KZT",
                 size="Amount_KZT", 
                 color="Risk_Score",
                 hover_name="Expense_Type", 
                 title="Internal Audit: Transaction Risk Analysis Profile",
                 labels={'Amount_KZT': 'Transaction Amount', 'Risk_Score': 'Risk Probability'},
                 color_continuous_scale=px.colors.sequential.Reds,
                 size_max=60)

# --- UI/UX Styling ---
# Applying professional dark theme for dashboard consistency
fig.update_layout(
    template='plotly_dark',
    title_x=0.5,
    font=dict(family="Arial, sans-serif", size=14)
)

# Render output
fig.show()
```

## 2. Interactive Anomaly Detection
Using Plotly to identify outliers by transaction amount and risk score.

```python
# --- Hierarchical Data Visualization (Treemap) ---
# Representing budget distribution and risk concentration across departments
fig_tree = px.treemap(df, 
                       path=['Department', 'Expense_Type'], 
                       values='Amount_KZT',
                       color='Risk_Score',
                       color_continuous_scale=px.colors.sequential.Reds,
                       title="Budget Allocation & Risk Density by Department")

fig_tree.update_layout(template='plotly_dark', title_x=0.5)

fig_tree.show()
```

```python
# --- Exporting Data for Tableau/BI tools ---
# Saving processed audit data to CSV for further visualization
df.to_csv('audit_data_for_tableau.csv', index=False)
print("File 'audit_data_for_tableau.csv' is ready for export!")
```
